```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

info = pd.read_csv("info.csv")
```

```python
stages = [1,2,3,4]
fig = plt.figure(figsize = (20,10))
gs = fig.add_gridspec(2,2)
(ax1,ax2),(ax3,ax4) = gs.subplots()
ax1.set_title("Pacman Evolution Stage 1")
ax3.set_title("Ghosts Evolution Stage 1")
ax2.set_title("Pacman Evolution Stage 2")
ax4.set_title("Ghosts Evolution Stage 2")
ax1.set_ylim([80,210])
ax3.set_ylim([0,300])
ax2.set_ylim([80,210])
ax4.set_ylim([0,300])
for test in range(50):
    y = []
    for stage in stages:
        thing = info[info["Stage"] == stage]
        thing = thing[thing["test"] == test]
        y.append(np.array(thing["Best_Fitness"]))
    ax1.plot(y[0])
    ax3.plot(y[1])
    ax2.plot(y[2])
    ax4.plot(y[3])
```

```python
stages = [1,2,3,4]
fig = plt.figure(figsize = (20,10))
gs = fig.add_gridspec(2,2)
(ax1,ax2),(ax3,ax4) = gs.subplots()
ax1.set_title("Pacman Evolution Stage 1")
ax3.set_title("Ghosts Evolution Stage 1")
ax2.set_title("Pacman Evolution Stage 2")
ax4.set_title("Ghosts Evolution Stage 2")
ax1.set_ylim([50,210])
ax3.set_ylim([0,300])
ax2.set_ylim([50,210])
ax4.set_ylim([0,300])

ys = []
for stage in stages:
    y = []
    for test in range(50):
        thing = info[info["Stage"] == stage]
        thing = thing[thing["test"] == test]
        y.append(np.array(thing["Best_Fitness"]))
    ys.append(y)

xs = []
xsu = []
xsd = []
for i in ys:
    x = []
    xu = []
    xd = []
    for j in i:
        x.append(np.mean(j))
        xu.append(np.mean(j) + np.std(j))
        xd.append(np.mean(j) - np.std(j))
    xs.append(x)
    xsu.append(xu)
    xsd.append(xd)
ax1.plot(xs[0])
ax1.fill_between(range(50), xsu[0], xsd[0], alpha=0.3)
ax3.plot(xs[1])
ax3.fill_between(range(50), xsu[1], xsd[1], alpha=0.3)
ax2.plot(xs[2])
ax2.fill_between(range(50), xsu[2], xsd[2], alpha=0.3)
ax4.plot(xs[3])
ax4.fill_between(range(50), xsu[3], xsd[3], alpha=0.3)
```
