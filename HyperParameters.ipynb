```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

info = pd.read_csv("data.csv")
```

```python
# using the info dataFrame, pull each combination of hyperparameters,
#     for each, subtract the best fitness of pacman in stage 3 from the best fitness of pacman in stage 1, 
#      and average those sums over the number of tests

# whichever has the highest average improvement will be considered optimal

# we should check to make sure the one that improves the most 
#  isn't just starting off the worst and getting to a still bad sum

popSizes = [100, 200, 500]
cProbs = [0.01, 0.05, 0.1]
mProbs = [0.01, 0.05, 0.1]
numGenerations = [25, 50, 100]

popSize = 50
probCrossover = 0.001
probMutate = 0.001
numGens = 2

bestps = 0
best2ps = 0
best3ps = 0
bestcp = 0
best2cp = 0
best3cp = 0
bestmp = 0
best2mp = 0
best3mp = 0
bestng = 0
best2ng = 0
best3ng = 0

bestImprov = 0
best2Improv = 0
best3Improv = 0
improvements = []

for popSize in popSizes:
    for probCrossover in cProbs:
        for probMutate in mProbs:
            for numGens in numGenerations:
                x = info[info["popSize"] == popSize]
                x = x[x["probCrossover"] == probCrossover]
                x = x[x["probMutate"] == probMutate]
                x = x[x["numGens"] == numGens]
                s1 = x[x["Stage"] == 1.]
                s1 = s1[s1["Generation"] == numGens - 1]
                s3 = x[x["Stage"] == 3.]
                s3 = s3[s3["Generation"] == numGens - 1]
                avgs = []
                top = np.array(s1["Best_Fitness"])
                bottom = np.array(s3["Best_Fitness"])
                for i in range(len(top)):
                    avgs.append(top[i] - bottom[i])
                    #print("")
                avg = np.mean(avgs)
                #print(popSize, probCrossover, probMutate, numGens, avg)
                if avg >= bestImprov:
                    best3Improv = best2Improv
                    best2Improv = bestImprov
                    bestImprov = avg
                    best3ps = best2ps
                    best2ps = bestps
                    bestps = popSize
                    best3cp = best2cp
                    best2cp = bestcp
                    bestcp = probCrossover
                    best3mp = best2mp
                    best2mp = bestmp
                    bestmp = probMutate
                    best3ng = best2ng
                    best2ng = bestng
                    bestng = numGens
                improvements.append(avg)

print("best Avg improvement:            " + str(bestImprov))
print("population size:                 " + str(bestps))
print("probability of crossover:        " + str(bestcp))
print("probability of mutate:           " + str(bestmp))
print("number of generations per stage: " + str(bestng))
print()
print("second best Avg improvement:     " + str(best2Improv))
print("population size:                 " + str(best2ps))
print("probability of crossover:        " + str(best2cp))
print("probability of mutate:           " + str(best2mp))
print("number of generations per stage: " + str(best2ng))
print()
print("third best Avg improvement:     " + str(best3Improv))
print("population size:                 " + str(best3ps))
print("probability of crossover:        " + str(best3cp))
print("probability of mutate:           " + str(best3mp))
print("number of generations per stage: " + str(best3ng))
```
