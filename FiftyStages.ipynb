```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

info = pd.read_csv("hundredEpochs.csv")
print(info)
```

```python
pacman = info[info["Team"] == 0]
ghosts = info[info["Team"] == 1]
```

```python
fig = plt.figure(figsize = (10,5))
for test in range(5):
    x = pacman[pacman["Test"] == test]
    x = x[x["Epoch"] == 0]
    x = np.array(x["Best_Fitness"])
    x = list(x)
    for epoch in range(1,50):
        y = pacman[pacman["Test"] == test]
        y = y[y["Epoch"] == epoch]
        y = np.array(y["Best_Fitness"])
        for i in y:
            x.append(i)
    plt.plot(x)
b = []
plt.ylim([50,300])
for i in range(len(x)):
    b.append(80)
plt.plot(b, label="score of 80")
plt.legend()
plt.xlabel("Generations")
plt.ylabel("Fitness")
```
